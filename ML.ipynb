```python
import pandas as pd

# Create dataset
data = {
    "Income": [400000, 30860, 133694, 149879, 140268],
    "Loan_Amount": [46606, 92313, 146699, 183792, 179073],
    "Credit_Score": [771, 362, 468, 798, 691],
    "DTI": [0.27, 0.65, 0.10, 0.35, 0.30],
    "Loan_Approved": [0, 0, 1, 0, 1]
}

df = pd.DataFrame(data)

# Split features and target
X = df.drop("Loan_Approved", axis=1)
y = df["Loan_Approved"]

# Train model/ Import Libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
#Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
#Train Model
model = LogisticRegression()
model.fit(X_train, y_train)

# Prediction
pred = model.predict(X_test)
print(pred)
#accuracey y test
accuracy = accuracy_score(y_test, pred)
print("accuracy:", accuracy)

#predict new customer
new_customer = [[500000, 600000, 700, 0.25]]
prediction = model.predict(new_customer)

print("Loan Approved:" , prediction)
```
